```python
# Import Pandas
import pandas as pd
```

```python
# Read data file
df = pd.read_csv("https://raw.githubusercontent.com/nhduy-bbean/MIS-301/refs/heads/main/01_Shopping_Smart_Dataset.csv")
df
```

## ***Basic Exploration***

### 1.How many items are listed in the dataset?

```python
# The number of items that are listed in the dataset
print("The number of items that are listed in the dataset is:")
len(df["item"])
```

### 2.Use .info() to check column types and if there are any missing values.

```python
# Check column types and if there are any missing values
df.info()
# There is no missing value
```

### 3.What is the average price in each supermarket?

```python
# The average price in each supermarket
three_supermarkets = df[["supermarket_A", "supermarket_B", "supermarket_C"]]
avg_price_each_market = three_supermarkets.mean()
avg_price_each_market
```

### 4.What’s the overall average price per item (across 3 supermarkets)?

```python
# The overall average price per item (across 3 supermarkets)
avg_price_per_item = three_supermarkets.mean(axis = 1)
df["avg_price_per_item"] = avg_price_per_item
df
```

### 5.What is the min, max, and range (max - min) price for each item?

```python
# Min, max, and range (max - min) price for each item
df["min_price_per_item"] = three_supermarkets.min(axis = 1)
df["max_price_per_item"] = three_supermarkets.max(axis = 1)
df["range_price_per_item"] = df["max_price_per_item"] - df["min_price_per_item"]
df
```

## ***Comparison & Logic***

### 6.Which supermarket offers the lowest average price across all items?

We will use the function idminx() on the variable avg_price_each_market the row label of the lowest value

```python
lowest_avg_price_supermarket = avg_price_each_market.idxmin()
print(lowest_avg_price_supermarket)
```

### 7.Create a new column cheapest_store that identifies the lowest priced store for each item.

.idxmin(axis = 1) means it will goes through the small dataframe of three_supermarkets (which only includes value of 3 markets' prices) and will find the label of which the data has the smallest value. and "axis = 1" means it will goes through every column of each row, which will find the market with the cheapest price.

```python
df["cheapest_store"] = three_supermarkets.idxmin(axis = 1)
df
```

### 8.Count how many times each supermarket is the cheapest.

The function value_counts() will return the number of frequencies in which the value occur.

```python
cheapest_counts = df["cheapest_store"].value_counts()
print(cheapest_counts)
```

### 9.Which item has the largest price difference among the three supermarkets?

With idxmax(), we can find the index of the data with the highest "range price per item". And then we will need to use loc[] to locate and gives information on that specific row.

```python
largest_difference_item = df.loc[df["range_price_per_item"].idxmax()]
print(largest_difference_item)
```

### 10.How many items have a price difference greater than 10,000 VND?

I will create a smaller dataframe that only includes data with the values in column "range_price_per_item" over 10000VND. And finally use len() to find how many items there are in that data frame.

```python
items_greater_than_10000 = df[df["range_price_per_item"] > 10000]
number_of_items_with_large_difference = len(items_greater_than_10000)
print(number_of_items_with_large_difference)
```

## ***Control Flows & Functions***

### 11.Write a function that returns "Expensive" if the item’s average price is over 80,000 VND.

```python
# Returns "Expensive" if the item's average price is over 80,000 VND
def check_price(price):
  if price > 80000:
    return "Expensive"
  else:
    return "Not Expensive"
```

### 12.Apply this function to create a column price_tag. How many items are marked "Expensive"?

```python
# Create a column price_tag
df["price_tag"] = df["avg_price_per_item"].apply(check_price)
df
```

```python
# Number of items are marked "Expensive"
df["price_tag"].value_counts()
# There are 113 items marked "Expensive"
```

### 13.Create a new column value_store that returns:
- 'A' if Supermarket A has the lowest price
- 'B' if B is lowest
- 'C' if C is lowest
Use if-elif-else logic.

Uses comparison to check for each row, if a row of an item == the row of the lowest item, then it returns the corresponding store name.

```python
def get_value_store(row):
  if row["supermarket_A"] == row[["supermarket_A", "supermarket_B", "supermarket_C"]].min():
    return 'A'
  elif row["supermarket_B"] == row[["supermarket_A", "supermarket_B", "supermarket_C"]].min():
    return 'B'
  else:
    return 'C'

df["value_store"] = df.apply(get_value_store, axis=1)
df
```

### 14.Write a function compare_prices(item_row) that prints the price difference between the highest and lowest store for that item.

We first will need to get prices of that argument from the 3 supermarkets, and use .max() and min() to find the largest and smallest price of the item needed in the argument to subtract them to find the range.

```python
def compare_prices(item_row):
  prices = item_row[["supermarket_A", "supermarket_B", "supermarket_C"]]
  price_difference = prices.max() - prices.min()
  print(item_row['item'],":",price_difference,"VND")
```

I will demonstrate the function on index 2 using .iloc[]

```python
compare_prices(df.iloc[2])

df.head()
```

### 15.Use a loop to apply that function to the first 5 items in the dataset.


i means index of the data, row is the variable that temporarily holds data during the loop.

```python
for i, row in df.head(5).iterrows():
  compare_prices(row)
```
